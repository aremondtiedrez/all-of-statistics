Let $X \sim N(0,\,1)$ and let $Y = e^X$.
1. Find the PDF for $Y$. Plot it.
2. Generate a vector $x = (x_1,\,\dots,\,x_{10,000})$ consisting of 10,000 random standard Normals.
Let $y = (y_1,\,\dots,\,y_{10,000})$ where $y_i = e^{x_i}$.
Draw a histogram of $y$ and compare it to the PDF you found in part 1.

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm
```

The PDF for $Y$ is given by $$f_Y(y) = \frac{1}{y\sqrt{2\pi}} e^{-(\ln^2 y) / 2}$$ and is plotted below.

```python
def f(y):
    return 1/(y*np.sqrt(2*np.pi)) * np.exp(-np.log(y)**2/2)

def plot_log_normal_pdf(xmax=5, num_pts=1000):

    x = np.linspace(0.01, xmax, num_pts) # Start the x values after zero to avoid division by zero
    y = f(x)

    plt.plot(x, y)
    plt.show()
```

```python
plot_log_normal_pdf(xmax=10, num_pts=1000)
```

We now sample 10,000 random standard normals, exponentiate them, and plot the resulting histogram.

```python
def compare_log_normal_hist_and_pdf(Y, xmax=5, num_pts=1000, num_bins=30):

    x = np.linspace(0.01, xmax, num_pts) # Start the x values after zero to avoid division by zero
    y = f(x)

    plt.hist(Y, bins=num_bins, density=True) # density=True normalises the histogram
    plt.plot(x, y)
    plt.xlim(0, xmax)
    plt.show()
```

```python
X = norm.rvs(size=1000000)
Y = np.exp(X)
compare_log_normal_hist_and_pdf(Y, xmax=3, num_bins=2000)
```

```python
compare_log_normal_hist_and_pdf(Y, xmax=8, num_bins=2000)
```
