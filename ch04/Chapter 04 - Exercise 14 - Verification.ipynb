#### Problem
Let $X_1,\,\dots,\,X_n \sim \text{Uniform}(0,1)$.
Let $Y_n = \overline{X}_n^2$ where $\overline{X}_n^2 = \frac{1}{n} \sum_{i=1}^n X_i$ denotes the sample mean.
Find the limiting distribution of $Y_n$.

#### Purpose of this notebook
We verify numerically the results obtained analytically, which are that (by the central limit theorem and the delta method)
$$
    Y_n \approx N \left( \frac{1}{4},\, \frac{1}{\sqrt{12}n} \right).
$$

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import uniform
```

```python
def generate_data(n, trials):
    
    X = uniform.rvs(size=n*trials).reshape([trials, n])
    sample_mean_X = np.mean(X, axis=1)
    
    Y = sample_mean_X**2
    
    sample_mean_Y = np.mean(Y)
    sample_var_Y = np.var(Y, ddof=1)
    
    return sample_mean_Y, sample_var_Y
```

```python
def plot_data(nmax, trials):
    
    n = np.arange(1, nmax + 1)
    
    data = np.transpose(np.array([generate_data(n=ni, trials=trials) for ni in n]))
    
    sample_mean_Y_data = data[0]
    sample_var_Y_data = data[1]
    
    plt.plot(n, sample_mean_Y_data)
    plt.axhline(y=0.25, color='r', linestyle='-')
    plt.show()

    plt.plot(n, sample_var_Y_data)
    plt.plot(n, 1/(12*n))
    plt.show()
```

```python
plot_data(nmax=50, trials=100)
```

```python
plot_data(nmax=100, trials=1000)
```
