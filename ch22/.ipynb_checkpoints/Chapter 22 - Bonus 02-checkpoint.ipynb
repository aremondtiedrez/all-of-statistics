In this notebook we use `sklearn` to implement *kernel* logistic regression.
We then apply it to the spam data.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy import stats
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import zero_one_loss
from sklearn.metrics.pairwise import pairwise_kernels
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

## Helper functions
Used to visualize two-dimensional binary classification samples and
the probability of classification into class $Y = 0$ given an estimator.

```python
def scatter_plot_two_classes(response, covariate, axis=None, alpha=0.3):
    """
    Given two arrays 'response' and 'covariate',
    where 'response' is a 1D array with shape (n, )
    whose values are all 0 or 1 and 'covariate'
    is a 2D array with shape (n, 2), plot
    the scatter plot of the values in 'covariate'
    coloured according to the corresponding
    values in 'response'.
    """

    # Verify that the arrays have appropriate sizes
    (ny, ) = response.shape
    (nx, dimension) = covariate.shape
    if dimension != 2:
        raise ValueError(
            "The rows of the argument 'covariate' "
            "must be pairs of real numbers."
        )
    if ny != nx:
        raise ValueError(
            "The arguments 'response' and 'covariate' "
            "must have the same number of rows."
        )
        
    # Class of samples where Y = 0
    class0 = covariate[response == 0]
    # Class of samples where Y = 1
    class1 = covariate[response == 1]

    # Create the plot
    if axis is None:
        figure, axis = plt.subplots(1)
    axis.set_aspect('equal')
    
    # Plot the sample points where Y = 0
    axis.scatter(class0[:, 0], class0[:, 1], color='#c1643e', alpha=alpha)
    # Plot the sample points where Y = 1
    axis.scatter(class1[:, 0], class1[:, 1], color='#3e9bc1', alpha=alpha)

def plot_predict_proba(
        fitted_model,
        xbound=None,
        x1min=0, x1max=1,
        x2min=0, x2max=1,
        N=100, # Number of grid points
        axis=None,
        levels=[0.1, 0.3, 0.5, 0.7, 0.9]
    ):
    """
    Given a fitted sklearn model trained on two-dimensional covariates,
    plot the probability of classifying (x1, x2) into class 0 according to that model.
    """

    # Create the points at which predict_proba will be evaluated
    if xbound is not None:
        x1min = x2min = -xbound
        x1max = x2max = xbound
    x1, x2 = np.meshgrid(np.linspace(x1min, x1max, N), np.linspace(x2min, x2max, N))
    # Evaluate predict_proba at every point
    # There are some shenanigans with np.reshape since x1 and x2 have shapes(N, N),
    # but we want them to have shape (K,) instead before passing them inside predict_proba.
    # Similarly: predict_proba evaluated on an array of shape (K,) returns an array
    # of shape (K,), but we need to reshape that array to (N, N) before it is passed into
    # plt.contourf
    z = np.reshape(fitted_model.predict_proba(np.column_stack((
        np.reshape(x1, (N**2, )),
        np.reshape(x2, (N**2, ))
    )))[:, 0], (N, N))

    # Create the plot
    if axis is None:
        figure, axis = plt.subplots()
    contour = axis.contour(x1, x2, z, levels=levels, cmap="coolwarm")
    
    # Include labels of the contour lines
    axis.clabel(contour)
```

# 1. Logistic Regression
First we implement logistic regression,
validating the implementation on a synthetic, two-dimensional, data set.

## Generate synthetic data

```python
n = 1000

# Covariate samples with Y = 0
X0 = np.column_stack((
    stats.uniform.rvs(loc=0, scale=3/4, size=n),
    stats.uniform.rvs(loc=0, scale=1, size=n)
))

# Covariate samples with Y = 1
X1 = np.column_stack((
    stats.uniform.rvs(loc=1/4, scale=3/4, size=n),
    stats.uniform.rvs(loc=0, scale=1, size=n)
))

# Full set of covariate samples
X = np.concatenate((X0, X1))

# Corresponding response samples
Y = np.concatenate((np.zeros(n), np.ones(n)))

# Plot the samples
scatter_plot_two_classes(Y, X)
```

## Perform logistic regression

```python
model = LogisticRegression()
fitted_model = model.fit(X, Y)

figure, axis = plt.subplots()
scatter_plot_two_classes(Y, X, axis=axis, alpha=0.15)
plot_predict_proba(fitted_model, axis=axis)
```

# 2. Kernel Logistic Regression
We now implement kernel logistic regression
validating the implementation on a synthetic, two-dimensional, data set.

## Generate synthetic data

```python
n = 1000

# Covariate samples with Y = 0
radius0 = stats.uniform.rvs(loc=4, scale=6, size=n)
angle0 = 2*np.pi*stats.uniform.rvs(loc=0, scale=1, size=n)
X0 = np.column_stack((
    radius0*np.cos(angle0),
    radius0*np.sin(angle0)
))

# Covariate samples with Y = 1
radius1 = stats.uniform.rvs(loc=6, scale=6, size=n)
angle1 = 2*np.pi*stats.uniform.rvs(loc=0, scale=1, size=n)
X1 = np.column_stack((
    radius1*np.cos(angle1),
    radius1*np.sin(angle1)
))

# Full set of covariate samples
X = np.concatenate((X0, X1))

# Corresponding response samples
Y = np.concatenate((np.zeros(n), np.ones(n)))

# Plot the samples
scatter_plot_two_classes(Y, X)
```

## Perform logistic regression
As expected, logistic regression performs very poorly on this synthetic data set
and assigns probability $1/2$ to every sample.

```python
model = LogisticRegression()
fitted_model = model.fit(X, Y)

figure, axis = plt.subplots()
scatter_plot_two_classes(Y, X, axis=axis, alpha=0.15)

plot_predict_proba(
    fitted_model, axis=axis,
    xbound=12,
    levels=[0.49, 0.5, 0.51]
)
```

## Define the KernelTransformer class

```python
class KernelTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self, kernel='rbf', **kernel_params):
        self.kernel = kernel
        self.kernel_params = kernel_params
        self.training_covariate = None

    def fit(self, covariate, response=None): 
        self.training_covariate = covariate
        return self

    def transform(self, covariate):
        return pairwise_kernels(
            covariate,
            self.training_covariate,
            metric=self.kernel,
            **self.kernel_params
        )
```

## Perform kernel logistic regression

### With a quadratic (polynomial) kernel

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("kernel", KernelTransformer(kernel='poly', coef0=1, degree=2)),
    ("logistic_regression", LogisticRegression())
])
fitted_pipeline = pipeline.fit(X, Y)

figure, axis = plt.subplots()
scatter_plot_two_classes(Y, X, axis=axis, alpha=0.15)

plot_predict_proba(
    fitted_pipeline, axis=axis,
    xbound=12
)
```

### With a Gaussian kernel

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("kernel", KernelTransformer(kernel='rbf')),
    ("logistic_regression", LogisticRegression(max_iter=1000))
])
fitted_pipeline = pipeline.fit(X, Y)

figure, axis = plt.subplots()
scatter_plot_two_classes(Y, X, axis=axis, alpha=0.15)

plot_predict_proba(
    fitted_pipeline, axis=axis,
    xbound=12
)
```

# Classify the spam data

## Download the spam data

```python
# Read the data into a pandas data frame
df = pd.read_csv('../data/spam.dat', sep=' ', header=None)

# Extract the response variable Y from the data frame
# and convert it to a numpy array
Y = df[df.columns[-1]].to_numpy()

# Extract all 57 covariates into a numpy array X
X = df[df.columns[:-1]].to_numpy()
```

## Use a quadratic (polynomial) kernel

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("kernel", KernelTransformer(kernel='poly', coef0=1, degree=2)),
    ("logistic_regression", LogisticRegression(max_iter=2000))
])
fitted_pipeline = pipeline.fit(X, Y)
empirical_error_rate = zero_one_loss(Y, fitted_pipeline.predict(X))
print(f"Empirical error rate: {empirical_error_rate:.3}")
```

## Use the Gaussian kernel

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("kernel", KernelTransformer(kernel='rbf')),
    ("logistic_regression", LogisticRegression(max_iter=1000))
])
fitted_pipeline = pipeline.fit(X, Y)
empirical_error_rate = zero_one_loss(Y, fitted_pipeline.predict(X))
true_error_rate_cv_estimate = 1 - cross_val_score(pipeline, X, Y, cv=5).mean()
print(f"Empirical error rate: {empirical_error_rate:.3}")
print(f"Cross-validation estimate of the true error rate: {true_error_rate_cv_estimate:.3}")
```
