Suppose a coin has probability $p$ of falling heads up.
If we flip the coin many times, we would expect the proportion of heads to be near $p$.
We will make this formal later.
Take $p=0.3$ and $n=1000$ and simulate $n$ coin flips.
Plot the proportion of heads as a function of $n$.
Repeat for $p = 0.03$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

def plot_coin_flips(p, n, ymin=0, ymax=1):

    # Flip the biased coin n times and compute the running proportion (running mean) of heads
    coinflips = bernoulli.rvs(p, size=n)
    runningsum = np.cumsum(coinflips)
    runningmean = [
        runsum/(idx + 1) for idx, runsum in enumerate(runningsum)
    ]
    
    # Plot
    plt.plot(np.linspace(1, n, n), runningmean)
    plt.plot(np.array([1, n]), np.array([p, p]))
    axes = plt.gca() # gca = Get Current Axes
    axes.set_ylim([ymin, ymax]) # Give the user the possibility to manually fix y-axis limits
    plt.show()
```

```python
plot_coin_flips(p=0.3, n=1000)
```

```python
plot_coin_flips(p=0.03, n=1000, ymax=0.06)
```
