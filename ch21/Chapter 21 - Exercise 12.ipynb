Repeat exercise 6 using the Haar basis.

```python
import matplotlib.pyplot as plt
import numpy as np

from collections import namedtuple
from scipy import integrate
```

Define the functions to expand along the Haar basis.

```python
t = [0.1, 0.13, 0.15, 0.23, 0.25, 0.40, 0.44, 0.65, 0.76, 0.78, 0.81]
h = (4, -5, 3, -4, 5, -4.2, 2.1, 4.3, -3.1, 2.1, -4.2)
step = lambda x: sum(
    hj*(1 + np.sign(x - tj))/2
    for tj, hj in zip(t, h)
)
doppler = lambda x: np.sqrt(x*(1-x))*np.sin(2.1*np.pi/(x + 1/20))
sine = lambda x: np.sin(np.pi*x)
cosine = lambda x: np.sqrt(2)*np.cos(3*np.pi*x)
```

Define the methods that compute the Haar basis coefficients
and reconstruct the approximation of the function given finitely
many of its coefficients.

```python
def scaling_coefficient(f):
    """
    Computes the scaling coefficient
    of the function f on [0, 1] with
    respect to the Haar basis.
    """

    return integrate.quad(f, 0, 1)[0]

def detail_coefficient(f, j, k):
    """
    Compute the detail coefficient,
    with respect to the Haar basis,
    of scale 2^{-j} and centered at
    (k + 1/2)/2^j, for the function
    f on [0, 1].
    """

    return (
        - 2**(j/2)*integrate.quad(f, k/2**j, (k + 1/2)/2**j)[0]
        + 2**(j/2)*integrate.quad(f, (k + 1/2)/2**j, (k+1)/2**j)[0]
    )

class Haar:

    def __init__(self, resolution):
        self.resolution = resolution
        self.scaling_coefficient = 0
        self.detail_coefficient = np.zeros((resolution + 1, 2**resolution))

    def analysis(self, f):
        """
        Compute the coefficients of
        the Haar basis expansion of
        the function f on [0, 1]
        up to j = self.resolution.
        """
    
        self.scaling_coefficient = scaling_coefficient(f)
        self.detail_coefficient = np.array(
            [[detail_coefficient(f, j, k) if k < 2**j else 0
            for k in range(2**self.resolution)]
            for j in range(self.resolution + 1)]
        )
        
    def synthesis(self, x_arr):
        """
        Reconstruct a function based on its
        Haar basis coefficients, then evaluate
        that function at the x-values in x_arr.
        """

        # Tile the array of detail coefficients
        # to create an array indexed by (x, j, k)
        # for x in x_arr
        tiled_detail_coefficients = np.tile(self.detail_coefficient, (len(x_arr), 1, 1))
        
        # Compute phi_{j, k} (x)
        x_mesh, j_mesh, k_mesh = np.meshgrid(
            x_arr,
            range(self.resolution + 1),
            range(2**self.resolution),
            indexing='ij'
        )
        wavelet_evaluations = 2**(j_mesh/2)*np.sign(2**j_mesh*x_mesh - k_mesh - 1/2)

        # Create the array beta_{j, k} * psi_{j, k} (x),
        # indexed by (x, j, k), keeping only the terms
        # where x is in the support of psi_{j, k}
        array_to_sum = np.where(
            np.abs(2**j_mesh*x_mesh - k_mesh - 1/2) <= 1/2, # Condition
            tiled_detail_coefficients*wavelet_evaluations, # Operation
            0 # Default value if condition fails
        )
        return self.scaling_coefficient + array_to_sum.sum(axis=(1, 2))
```

Compute the wavelet coefficients of the various functions
and plot their reconstructions.

```python
x_arr = np.linspace(0, 1, 1000)

figure, axes = plt.subplots(4)
for axis, (name, function, resolution) in zip(axes, (
    ('sine', sine, 5),
    ('cosine', cosine, 5),
    ('step', step, 6),
    ('doppler', doppler, 8),
)):
    haar_decomposition = Haar(resolution)
    haar_decomposition.analysis(function)
    y_arr = haar_decomposition.synthesis(x_arr)
    axis.plot(x_arr, function(x_arr))
    axis.plot(x_arr, y_arr, label=f"{name.capitalize()} (J = {resolution})")
    axis.legend(loc='best')
plt.show()
plt.close()
```

Verify the analytic expressions found for
the wavelet coefficients of the sine and cosine
functions.
This is done by plotting the reconstruction from
these coefficients against the true functions
(but we could also diretly compare the analytic value
of the coefficients to their numerically-obtained value).

First we define the analytic values of
the wavelet coefficients of the sine and cosine functions
and use them to create Haar objects.

```python
resolution = 6

sine_scaling_coefficient = 2/np.pi
def sine_detail_coefficient(j, k):
    return (
        (2**(j/2)/np.pi)*(
            - np.cos(np.pi*k/2**j)
            + 2*np.cos(np.pi*(k+1/2)/2**j)
            - np.cos(np.pi*(k+1)/2**j)
        )
    )
    
sine_haar = Haar(resolution)
sine_haar.scaling_coefficient = sine_scaling_coefficient
sine_haar.detail_coefficient = np.array(
    [[sine_detail_coefficient(j, k) if k < 2**j else 0
    for k in range(2**sine_haar.resolution)]
    for j in range(sine_haar.resolution + 1)]
)

cosine_scaling_coefficient = 0
def cosine_detail_coefficient(j, k):
    return (
        (2**(j/2)*np.sqrt(2)/(3*np.pi))*(
            np.sin(3*np.pi*k/2**j)
            - 2*np.sin(3*np.pi*(k+1/2)/2**j)
            + np.sin(3*np.pi*(k+1)/2**j)
        )
    )
    
cosine_haar = Haar(resolution)
cosine_haar.scaling_coefficient = cosine_scaling_coefficient
cosine_haar.detail_coefficient = np.array(
    [[cosine_detail_coefficient(j, k) if k < 2**j else 0
    for k in range(2**cosine_haar.resolution)]
    for j in range(cosine_haar.resolution + 1)]
)
```

Now we plot the reconstructions against the true functions.

```python
figure, axes = plt.subplots(2)
for axis, (name, function, haar_decomposition) in zip(axes, (
    ('sine', sine, sine_haar),
    ('cosine', cosine, cosine_haar),
)):
    y_arr = haar_decomposition.synthesis(x_arr)
    axis.plot(x_arr, function(x_arr))
    axis.plot(x_arr, y_arr, label=f"{name.capitalize()} (J = {resolution})")
    axis.legend(loc='best')
plt.show()
plt.close()
```

## Bonus
A dyadic step function, which is a (finite) linear combination
of Haar wavelets.

```python
t_dyadic = [0, 1/8, 1/4, 1/2]
h_dyadic = (1, -3, 1, 3)
dyadic_step = lambda x: sum(
    hj*(1 + np.sign(x - tj))/2
    for tj, hj in zip(t_dyadic, h_dyadic)
)

x_arr = np.linspace(0, 1, 1000)

# Decompose the function at resolutions J=1 and J=2
haar_decompositions = (Haar(resolution=1), Haar(resolution=2), Haar(resolution=3))
for i, haar_decomposition in enumerate(haar_decompositions):
    haar_decomposition.analysis(dyadic_step)
    y_arr = haar_decomposition.synthesis(x_arr)
    plt.plot(x_arr, y_arr, label=f"J = {i+1}")
plt.plot(x_arr, dyadic_step(x_arr), label='True function', linestyle='dashed')
plt.legend(loc='best')
plt.show()
plt.close()
```
