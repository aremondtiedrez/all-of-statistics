Consider again the Doppler signal
$$
    f(x) = \sqrt{x(1-x)} \sin \left( \frac{2.1\pi}{x + \frac{1}{20}} \right).
$$
Let $n = 1,024$, $\sigma = 0.1$, and let $(x_1, ..., x_n) = (1/n, ..., 1)$.
Generate data
$$
    Y_i = f(x_i) + \varepsilon_i
$$
where $\varepsilon_i \sim N(0, \sigma)$.
1. Fit the curve using the cosine basis method.
   Plot the function estimate and confidence band for $J = 10, 20, ..., 100$.
2. Use Haar wavelets to fit the curve.

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import stats

import import_ipynb
import cosineNB
import haarNB
```

```python
# Generating the data
doppler = lambda x: np.sqrt(x*(1-x))*np.sin(2.1*np.pi/(x + 1/20))

n = 1024
X = np.linspace(1/n, 1, n)
error = stats.norm.rvs(scale=0.1, size=n)
Y = doppler(X) + error

cosine_regression = cosineNB.CosineRegression(Y, X)
haar_regression = haarNB.HaarRegression(Y, X)

cosine_regression.plot(true_regression_function=doppler)
haar_regression.plot(true_regression_function=doppler)
```
