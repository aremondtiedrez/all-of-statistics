Expand the following functions in the cosine basis on $[0, 1]$.
1. $f(x) = \sum_{j=1}^{11} h_j K(x - t_j)$ where
   $$
        K(t)
        = \frac{1 + \text{sign}\; t}{2}
        = \left\{
        \begin{aligned}
            &0   &&\text{if } t < 0,\\
            &1/2 &&\text{if } t = 0, \text{ and }\\
            &1   &&\text{if } t > 0
        \end{aligned}
        \right.
   $$
   and where
   $$
        t = (0.1, 0.13, 0.15, 0.23, 0.25, 0.40, 0.44, 0.65, 0.76, 0.78, 0.81)
   $$
   and
   $$
        h = (4, -5, 3, -4, 5, -4.2, 2.1, 4.3, -3.1, 2.1, -4.2).
   $$
1. $f(x) = \sqrt{x(1-x)} \sin \left( \frac{2.1 \pi}{x + \frac{1}{20}} \right)$,
   which is known as the "Doppler function".

Then plot the partial sum
$$
    \sum_{j=1}^n \beta_j \phi_j (x)
$$
for increasing values of $n$.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

Define the functions to expand along the cosine basis.

```python
t = [0.1, 0.13, 0.15, 0.23, 0.25, 0.40, 0.44, 0.65, 0.76, 0.78, 0.81]
h = (4, -5, 3, -4, 5, -4.2, 2.1, 4.3, -3.1, 2.1, -4.2)
step = lambda x: sum(
    hj*(1 + np.sign(x - tj))/2
    for tj, hj in zip(t, h)
)
doppler = lambda x: np.sqrt(x*(1-x))*np.sin(2.1*np.pi/(x + 1/20))
```

Plot these functions.

```python
x_arr = np.linspace(0, 1, 1000)

figure, axes = plt.subplots(2)
axes[0].plot(x_arr, step(x_arr))
axes[1].plot(x_arr, doppler(x_arr))
plt.show()
plt.close()
```

Define the methods that compute the cosine basis coefficients
and reconstruct the approximation of the function given finitely
many of its coefficients.

```python
def cosine_coefficient(f, j, t=None):
    """
    Computes the j-th cosine basis coefficient
    of the function f on [0, 1].

    If t is provided then the integrals are
    computed after first subdividing [0, 1] into
    [0, t[0]], [t[0], t[1]], ..., [t[-1], 1].
    """

    if t is None:
        t = []
    t_tilde = [0] + t + [1]

    if j==0:
        return sum(
            scipy.integrate.quad(f, t_tilde[i], t_tilde[i+1])[0]
            for i in range(len(t) + 1)
        )

    return np.sqrt(2)*sum(
        scipy.integrate.quad(f, t_tilde[i], t_tilde[i+1], weight='cos', wvar=j*np.pi)[0]
        for i in range(len(t) + 1)
    )

def cosine_analysis(f, n, t=None):
    """
    Computes the first n cosine basis coefficients
    of the function f on [0, 1].

    If t is provided then the integrals are
    computed after first subdividing [0, 1] into
    [0, t[0]], [t[0], t[1]], ..., [t[-1], 1].
    """

    return np.array([cosine_coefficient(f, j, t) for j in range(n + 1)])

def cosine_synthesis(coefficients, x_arr):
    """
    Reconstructs a function based on its cosine basis
    coefficients, ordered from j=0 to n, then
    evaluates that function at the x-values in x_arr.
    """

    j_mesh, x_mesh = np.meshgrid(range(1, len(coefficients)), x_arr, indexing='ij')
    coeff_mesh = coefficients[j_mesh]
    y_arr = coefficients[0] + np.sqrt(2)*np.sum(
        coeff_mesh*np.cos(np.pi*j_mesh*x_mesh),
        axis = 0
    )

    return y_arr
```

Compute the first 300 coefficients of the step function $f$.

```python
n_max = 300
step_coefficients = cosine_analysis(step, n_max, t)
```

Plot successive approximations of the step function $f$.

```python
n_list = [10, 50, 300]
x_arr = np.linspace(0, 1, 1000)

figure, axes = plt.subplots(3)
for axis, n in zip(axes, n_list):
    y_arr = cosine_synthesis(step_coefficients[:n], x_arr)
    axis.plot(x_arr, step(x_arr))
    axis.plot(x_arr, y_arr, label=f"n = {n}")
    axis.legend(loc="best")
plt.show()
plt.close()
```

Compute the first 1000 coefficients of the doppler function.

```python
n_max = 1000
doppler_coefficients = cosine_analysis(doppler, n_max)
```

Plot successive approximations of the doppler function.

```python
n_list = [10, 100, 1000]
x_arr = np.linspace(0, 1, 1000)

figure, axes = plt.subplots(3)
for axis, n in zip(axes, n_list):
    y_arr = cosine_synthesis(doppler_coefficients[:n], x_arr)
    axis.plot(x_arr, doppler(x_arr))
    axis.plot(x_arr, y_arr, label=f"n = {n}")
    axis.legend(loc="best")
plt.show()
plt.close()
```

Finally we consider the sine function $f(x) = \sin(\pi x)$.
Its coefficients along the cosine basis were computed analytically to be
$$
    \beta_0 = \frac{2}{\pi},
$$
$\beta_j = 0$ when $j$ is odd, and
$$
    \beta_j = \frac{2\sqrt{2}}{\pi(1 - j^2)}
$$
when $j \geqslant 2$ is even.
Below we verify numerically that inverting these coefficients approaches the sine function.

Define the sine function and its coefficients.

```python
sine = lambda x: np.sin(np.pi*x)

def sine_coefficient(j):
    if j%2 == 1:
        return 0
    elif j == 0:
        return 2/np.pi
    return 2*np.sqrt(2)/(np.pi*(1 - j**2))
```

Compute the first 5000 coefficients of the sine function.

```python
n_max = 5000
sine_coefficients = np.array([sine_coefficient(j) for j in range(n_max + 1)])
```

Plot successive approximations of the sine function.

```python
n_list = [5, 10, 100, 5000]
x_arr = np.linspace(0, 1, 1000)

figure, axes = plt.subplots(4)
for axis, n in zip(axes, n_list):
    y_arr = cosine_synthesis(sine_coefficients[:n], x_arr)
    axis.plot(x_arr, sine(x_arr))
    axis.plot(x_arr, y_arr, label=f"n = {n}")
    axis.legend(loc="best")
plt.show()
plt.close()
```
