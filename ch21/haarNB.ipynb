Consider the glass fragments data from the book's website.
Let $Y$ be the refractive index and let $X$ be the alumninum content
(the fourth variable).
1. Do a nonparametric regression to fit the model
   $Y = f(x) + \varepsilon$ using the cosine basis method.
   The data are not on a regular grid.
   Ignore this when estimating the function.
   (But do sort the data first according to $X$.)
   Provide a function estimate, an estimate of the risk, and a confidence band.
1. Use the wavelet method to estimate $f$.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy import stats
```

## Download the glass fragment data

```python
# Read the data into a pandas data frame
df = pd.read_csv('../data/glass_clean.dat', sep='\t')

# Transform the two dataframe columns of interest into an array
# where the first column corresponds to X,
# the second column corresponds to Y, and
# each row corresponds to a sample.
glass_X = df['Al'].to_numpy()
glass_Y = df['RI'].to_numpy()
```

## The HaarRegression class

```python
class HaarRegression():

    def __init__(self,
        response,
        covariate,
        x_arr = np.linspace(0, 1, 1000),
        alpha = 0.05
    ):
        """
        Initializes a HaarRegression object given two 1D arrays
        response and covariate.
        """

        self.number_of_samples = len(response)
        self.response = response
        self.covariate = covariate
        # The x-values in x_arr must be in [0, 1]
        self.x_arr = x_arr
        # Normalize the covariate data, if necessary
        if covariate.min() < 0 or covariate.max() > 1:
            self.covariate_normalized = (
                (covariate - covariate.min())/(covariate.max() - covariate.min())
            )
            self.x_arr_scaled = (
                (self.covariate.max() - self.covariate.min())
                *self.x_arr + self.covariate.min()
            )
        else:
            self.covariate_normalized = covariate
            self.x_arr_scaled = x_arr
        self.resolution = int(np.floor( np.log2(self.number_of_samples) )) - 1
        self.scaling_coefficient = None
        self.detail_coefficients = None
        self.is_tresholding_done = False
        self.variance_estimate = None
        self.alpha = alpha
        self.confidence_band = None

    def compute_coefficients(self):
        """
        Computes the empirical estimates of the Haar
        wavelet scaling and detail coefficients of
        the regression function, given data.

        Here we use the normalized version of the covariates.
        """

        # The scaling coefficient is the sample mean of the response
        self.scaling_coefficient = np.mean(self.response)

        # We now compute the detail coefficients
        # First we compute the 3D array psi_{j, k} (X_i).
        # which we call wavelet_evaluation_sample
        xi_mesh, j_mesh, k_mesh = np.meshgrid(
            self.covariate_normalized,
            range(self.resolution + 1),
            range(2**self.resolution),
            indexing='ij'
        )
        wavelet_evaluation_sample = np.where(
            (k_mesh <= 2**j_mesh - 1)
            & (np.abs(2**j_mesh*xi_mesh - k_mesh - 1/2) <= 1/2),
            2**(j_mesh/2)*np.sign(2**j_mesh*xi_mesh - k_mesh - 1/2),
            0
        )
        # We now compute the 3D array Y_i psi_{j, k} (X_i),
        # which we call detail_coefficient_sample
        detail_coefficient_sample = np.einsum(
            'i, ijk -> ijk',
            self.response,
            wavelet_evaluation_sample
        )
        # Finally we compute the details coefficients,
        # indexing them by (j, k).
        # Note that k ranges all the way up to 2^J - 1,
        # not just 2^j - 1, in order to work with
        # a rectangular array.
        self.detail_coefficients = np.mean(detail_coefficient_sample, axis=0)

    def compute_variance_estimate(self):
        """
        Compute the variance estimate used in universal tresholding.
        """

        # If necessary, compute the detail coefficients
        if self.detail_coefficients is None:
            self.compute_coefficients()
            
        self.variance_estimate = (
            np.median(np.abs(self.detail_coefficients[-1]))
            /stats.norm.ppf(3/4)
        )

    def perform_tresholding(self):
        """
        Perform the universal tresholding where
        all detail coefficients which are, in
        absolute value, sufficiently small are
        zeroed out.
        """

        # If necessary, compute the detail coefficients
        if self.detail_coefficients is None:
            self.compute_coefficients()
        # If necesary, compute the variance estimate
        if self.variance_estimate is None:
            self.compute_variance_estimate()

        # Evaluate the treshold value
        treshold_value = self.variance_estimate*np.sqrt(
            2*np.log(self.number_of_samples)#/self.number_of_samples
        )

        # Treshold the detail_coefficients
        self.detail_coefficients = np.where(
            np.abs(self.detail_coefficients) > treshold_value,
            self.detail_coefficients,
            0
        )
        self.is_tresholding_done = True

    def plot(self, axis=None, show_scatter_plot=True, true_regression_function=None):
        """
        Given a matplotlib.pyplot axis object,
        and possibly the true regression function,
        plot an evaluation of the Haar regression
        estimator at the x-values of self.x_arr_scaled.
        """

        # If axis is not provided, create it
        if axis is None:
            figure, axis = plt.subplots(1)

        y_arr = self.synthesis()
        axis.plot(
            self.x_arr_scaled,
            y_arr,
            color='crimson',
            label='Regression estimator'
        )

        # For synthetic data: if the true regression
        # function is provided, plot it as well
        if true_regression_function is not None:
            axis.plot(
                self.x_arr_scaled,
                true_regression_function(self.x_arr_scaled),
                label='True regression function'
            )

        # If asked, show a scatter plot of the oririginal data
        if show_scatter_plot:
            axis.scatter(
                self.covariate,
                self.response, 
                color='darkseagreen',
                alpha=0.3,
                label='Data'
            )

        axis.legend(loc='best')

    def synthesis(self):
        """
        Reconstruct the Haar wavelet regression estimator
        from its scaling and detail coefficients and
        evaluate that estimator at the x-values in self.x_arr.

        The x-values in self.x_arr must be in [0, 1].

        Returns
        y_arr       1D array containing the values of
                    the estimator evaluated at the x-values
                    in self.x_arr.
        """

        # If necessary, compute the Haar wavelet coefficients
        if self.scaling_coefficient is None or self.detail_coefficients is None:
            self.compute_coefficients()
        # If necessary, treshold the Haar wavelet coefficients
        if not self.is_tresholding_done:
            self.perform_tresholding()

        # Tile the array of detail coefficients
        # to create an array indexed by (x, j, k)
        # for x in x_arr
        tiled_detail_coefficients = np.tile(self.detail_coefficients, (len(self.x_arr), 1, 1))

        # Compute psi_{j, k} (x)
        x_mesh, j_mesh, k_mesh = np.meshgrid(
            self.x_arr,
            range(self.resolution + 1),
            range(2**self.resolution),
            indexing='ij'
        )
        wavelet_evaluations = 2**(j_mesh/2)*np.sign(2**j_mesh*x_mesh - k_mesh - 1/2)

        # Create the array beta_{j, k} * psi_{j, k} (x),
        # indexed by (x, j, k), keeping only the terms
        # where x is in the support of psi_{j, k}
        array_to_sum = np.where(
            (k_mesh <= 2**j_mesh - 1) # Conditions
            & (np.abs(2**j_mesh*x_mesh - k_mesh - 1/2) <= 1/2),
            tiled_detail_coefficients*wavelet_evaluations, # Operation
            0 # Default value if condition fails
        )
        
        y_arr = self.scaling_coefficient + array_to_sum.sum(axis=(1, 2))

        return y_arr
```

```python
glass_haar_regression = HaarRegression(glass_Y, glass_X)
glass_haar_regression.plot()
```

## Testing the HaarRegression class
We test the Haar class by using it in on synthetic data
generated from a variety of regression function.

```python
# Regressions functions we will use
poly = lambda x: x*(1-x)*x**3
trig = lambda x: np.cos(2*np.pi*x)**3
# This dyadic step function is an exact
# linear combination of Haar wavelets
t = [0, 1/8, 1/4, 1/2]
h = (1, -3, 1, 3)
dyadic_step = lambda x: sum(
    hj*(1 + np.sign(x - tj))/2
    for tj, hj in zip(t, h)
)

figure, axes = plt.subplots(3)
figure.set_figheight(12)

"""
We perform Haar wavelet regression on several synthetic datasets.

Parameters
resolution          Value of the resolution J to use.
error_std_dev       Standard deviation of the error term
                    added to f(X) to produce Y.
[a, b]              The interval on which we uniformly
                    sample the covariate X.
"""

# err
for axis, (function, number_of_samples, error_std_dev, J, a, b) in zip(axes, (
    (dyadic_step, 2**8 , 0.5,  2, 0, 1),
    (poly,        2**10, 0.01, 4, 0, 1),
    (trig,        2**10, 0.1,  6,-2, 5)
)):
    covariate = stats.uniform.rvs(loc=a, scale=b-a, size=number_of_samples)
    error = stats.norm.rvs(scale=error_std_dev, size=number_of_samples)
    response = function(covariate) + error
    haar_regression = HaarRegression(response, covariate)
    haar_regression.resolution = J
    haar_regression.plot(axis=axis, true_regression_function=function)
plt.show()
```

```python
doppler_function = lambda x: np.sqrt(x*(1-x))*np.sin(2.1*np.pi/(x + 1/20))

n = 2**11 # 2**10 = 1024
doppler_X = stats.uniform.rvs(size=n)
error = stats.norm.rvs(scale=0.1, size=n)
doppler_Y = doppler_function(doppler_X) + error

# For now we perform the 'tresholding' manually
# by actually **truncating** at J = 3
doppler_haar_regression = HaarRegression(doppler_Y, doppler_X)
doppler_haar_regression.resolution = 6

# Compute the Haar wavelet coefficients
doppler_haar_regression.compute_coefficients()

# Plot the Haar regression estimator
doppler_haar_regression.plot(true_regression_function=doppler_function)

print(doppler_haar_regression.detail_coefficients[-1])
print(doppler_haar_regression.is_tresholding_done)
print(doppler_haar_regression.scaling_coefficient)
```
