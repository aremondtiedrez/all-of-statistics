# Haar density estimator
We construct the Haar density estimator for a $Beta(14, 5)$ distribution
using $n = 1,000$ samples.

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import stats
```

```python
class HaarDensityEstimation():

    def __init__(self,
        data,
        x_arr = np.linspace(0, 1, 1000),
    ):
        """
        Initializes a HaarDensityEstimation object given
        a 1D array containing the data points.
        """

        self.number_of_samples = len(data)
        self.data = data
        # We only consider resolutions up to 2**J = sqrt(n)
        self.resolution = int(np.floor(np.log2(self.number_of_samples)/2))
        self.detail_coefficients = None
        self.variance_estimates = None
        self.x_arr = x_arr
        self.y_arr = None

    def compute_coefficients_and_variance(self):
        """
        Computes the empirical esttimates of the Haar
        wavelet detail coefficients of the density, given data.
        """

        # We compute the 3D array psi_{j, k} (X_i).
        # which we call wavelet_evaluation_sample
        xi_mesh, j_mesh, k_mesh = np.meshgrid(
            self.data,
            range(self.resolution + 1),
            range(2**self.resolution),
            indexing='ij'
        )
        wavelet_evaluation_sample = np.where(
            (k_mesh <= 2**j_mesh - 1)
            & (np.abs(2**j_mesh*xi_mesh - k_mesh - 1/2) <= 1/2),
            2**(j_mesh/2)*np.sign(2**j_mesh*xi_mesh - k_mesh - 1/2),
            0
        )
        
        # We compute the details coefficients,
        # as well as the variance estimates,
        # indexing them by (j, k).
        # Note that k ranges all the way up to 2^J - 1,
        # not just 2^j - 1, in order to work with
        # a rectangular array.
        self.detail_coefficients = np.mean(wavelet_evaluation_sample, axis=0)
        self.variance_estimates = np.var(wavelet_evaluation_sample, axis=0, ddof=1)

    def compute_resolution_from_risk_estimate(self):
        """
        Compute the risk estimate for B = 0, ..., log_2(n)
        then set the resolution to the value of B that
        minimizes it.
        """

        # If necesary, compute the detail coefficients
        # and the variance estimates
        if self.detail_coefficients is None or self.variance_estimates is None:
            self.compute_coefficients_and_variance()

        # We compute the contribution to the risk 
        # estimate coming from the variance
        variance_contribution = (
            np.cumsum(np.sum(self.variance_estimates, axis=1))
            /self.number_of_samples
        )

        # We compute the contribution to the risk
        # estimate coming from the bias
        bias_estimates = (
            self.detail_coefficients**2 - self.variance_estimates/self.number_of_samples
        )
        # Take the positive part
        bias_estimates = np.where(bias_estimates > 0, bias_estimates, 0)
        # Sum the bias estimates to produce the bias contribution
        bias_contribution = np.cumsum(np.sum(self.variance_estimates, axis=1)[::-1])[::-1]

        # Combine the contributions from the variance
        # and from the bias to produce the risk estimate
        risk_estimate = variance_contribution + bias_contribution

        # Choose the resolution to minimize the risk estimate
        self.resolution = np.argmin(risk_estimate)

    def plot(self, true_density=None):
        """
        Possible given the true density, plot
        an evaluation of the Haar histogram at
        the x_values of self.x_arr.
        """

        # If necessary, evaluate the Haar histogram
        # at the x-values of x_arr
        if self.y_arr is None:
            self.evaluate()

        # The Haar histogram
        plt.plot(self.x_arr, self.y_arr, color='crimson', label='Haar histogram')

        # For synthetic data: if the true density
        # is provided, plot it as well
        if true_density is not None:
            plt.plot(
                self.x_arr,
                true_density(self.x_arr),
                color='darkseagreen',
                label='True density'
            )

        plt.legend(loc='best')

    def evaluate(self):
        """
        Reconstruct the Haar histogram from its detail
        coefficients (the scaling coefficient is always
        equal to 1) and evaluate it at the x-values
        in self.x_arr.
        """

        # If necessary, compute the Haar wavelet coefficients
        if self.detail_coefficients is None:
            self.compute_coefficients_and_variance()

        # Tile the array of detail coefficients
        # to create an array indexed by (x, j, k)
        # for x in x_arr
        tiled_detail_coefficients = np.tile(
            self.detail_coefficients,
            (len(self.x_arr), 1, 1)
        )

        # Compute psi_{j, k} (x)
        x_mesh, j_mesh, k_mesh = np.meshgrid(
            self.x_arr,
            range(self.resolution + 1),
            range(2**self.resolution),
            indexing='ij'
        )
        wavelet_evaluations = 2**(j_mesh/2)*np.sign(2**j_mesh*x_mesh - k_mesh - 1/2)

        # Create the array beta_{j, k} * psi_{j, k} (x),
        # indexed by (x, j, k), keeping only the terms
        # where x is in the support of psi_{j, k}
        array_to_sum = np.where(
            (k_mesh <= 2**j_mesh - 1) # Conditions
            & (np.abs(2**j_mesh*x_mesh - k_mesh - 1/2) <= 1/2),
            tiled_detail_coefficients*wavelet_evaluations, # Operation
            0 # Default value if condition fails
        )

        scaling_coefficient = 1
        self.y_arr = scaling_coefficient + array_to_sum.sum(axis=(1, 2))
```

```python
# Generate the data
n = 1000
X = stats.beta.rvs(a=14, b=5, size=n)

# The true density (used for plotting later)
beta_pdf = lambda x: stats.beta.pdf(x, a=14, b=5)

# Perform the Haar density estimation,
# choosing the resolution which minimizes
# the risk estimate.
haar = HaarDensityEstimation(X)
initial_resolution = haar.resolution
haar.compute_resolution_from_risk_estimate()
optimal_resolution = haar.resolution

# Report out whether or not optimizing
# the resolutio selected a smaller resolution
if initial_resolution == optimal_resolution:
    print("The maximal resolution is the optimal resolution.")
else:
    print("The optimal resolution is COARSER than the maximal resolution.")

# Plot the Haar histogram alongside
# the true density
haar.plot(true_density=beta_pdf)
plt.show()
plt.close()
```
