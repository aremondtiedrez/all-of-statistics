```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
# Auxiliary functions when n = 1

def wald_pvalue(x):
    return 2*scipy.stats.norm.cdf(-np.abs(x))

def bayes_prob_null(x, b):
    
    tau = np.sqrt( b**2 / (1 + b**2) )
    mu = tau**2 * x
    
    prob = 1 / (1 + (tau / b) * np.exp( mu**2 / (2 * tau**2) ))
    
    return prob
```

```python
# Plotting function when n = 1

def plot_comparison(b):
    
    x = np.linspace(-3, 3, 1000)

    y_pvalue = wald_pvalue(x)
    y_bayes = bayes_prob_null(x, b)
    y_baseline = [0.5 for xi in x]

    plt.plot(x, y_pvalue, label='p-value of the Wald test')
    plt.plot(x, y_bayes, label=f'Bayesian probability of the null (b = {b})')
    plt.plot(x, y_baseline)
    plt.legend(loc='best')
    plt.xlabel('Value of X')
    
    plt.show()
```

```python
# A few plots corresponding to n = 1

b_list = (
    0.1,
    1,
    10
)

for b in b_list:
    plot_comparison(b)
```

```python
# Auxiliary functions when n > 1

def wald_pvalue_n(X):
    
    n = len(X)
    
    pval = 2*scipy.stats.norm.cdf(
        -np.sqrt(n)
        *np.abs(X.mean())
    )
    
    return pval

def bayes_prob_null_n(X, b):
    
    n = len(X)
    
    tau = np.sqrt( b**2 / (1 + n * b**2) )
    mu = n * tau**2 * X.mean()
    
    prob = 1 / (1 + (tau / b) * np.exp( mu**2 / (2 * tau**2) ))
    
    return prob
```

```python
# Plotting function when n > 1

def compare_and_plot(true_mu, n, b_log_scale_start, b_log_scale_stop, want_to_plot=True):
    
    # Generate data under the ALTERNATIVE hypothesis
    # (to ensure that the pvalue be small)
    X = scipy.stats.norm.rvs(loc=true_mu, size=n)

    # Plot, for a range of vales of the prior variance b,
    # the Bayesian posterior probability of the null hypothesis
    b_plot = np.logspace(start=b_log_scale_start, stop=b_log_scale_stop, num=1000)

    y_bayes = [bayes_prob_null_n(X, b) for b in b_plot]

    if want_to_plot:
        plt.plot(b_plot, y_bayes, label='Bayesian probability of the null')
        plt.xscale('log')
        plt.legend(loc='best')
        plt.xlabel('Prior variance b')

        plt.show()
        
    return y_bayes
```

```python
# Plot after a SINGLE experiment

bayes = compare_and_plot(
    true_mu=0.03,
    n=int(1e4),
    b_log_scale_start=-5,
    b_log_scale_stop=7,
    want_to_plot=True
)
```

```python
def bayesian_average_plot(
    trials,
    true_mu,
    n,
    b_log_scale_start,
    b_log_scale_stop
):

    """
    Perform many trials, then average the
    Bayesian probability of the null
    obtained in each case, then plot it.
    This shows an estimation of the true
    Bayesian probability of the null
    for various values of the prior variance b.
    """
    
    bayes_list = np.array([
        compare_and_plot(
            true_mu,
            n,
            b_log_scale_start,
            b_log_scale_stop,
            want_to_plot=False
        )
        for _ in range(trials)
    ])

    b_plot = np.logspace(start=b_log_scale_start, stop=b_log_scale_stop, num=1000)
    avg_bayes = bayes_list.mean(axis=0)

    plt.plot(b_plot, avg_bayes, label='Bayesian probability of the null\naveraged over many trials')
    plt.xscale('log')
    plt.legend(loc='best')
    plt.xlabel('Prior variance b')

    plt.show()
```

```python
bayesian_average_plot(
    trials=500,
    true_mu=0.03,
    n=int(1e4),
    b_log_scale_start=-5,
    b_log_scale_stop=7
)
```
