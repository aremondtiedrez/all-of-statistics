```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
def plot_beta_posterior(a, b, s, n):
    """
    Given a Beta(a, b) prior,
    we observe n Bernoulli tosses
    with s successes.
    
    We plot the prior and
    the Beta(a + s, b + n - s)
    posterior.
    """
    
    x = np.linspace(0, 1, 1001)
    
    prior = scipy.stats.beta.pdf(x, a, b) 
    posterior = scipy.stats.beta.pdf(x, a = a + s, b = b + n - s)

    plt.plot(x, prior, label=f'prior (a={a}, b={b})')
    plt.plot(x, posterior, label=f'prior (a={a+s}, b={b+n-s})')

    plt.legend(loc='best')
    plt.show()
```

```python
s = 2
n = 10

prior_parameters = (
    (0.5, 0.5),
    (1, 1),
    (10, 10),
    (100, 100)
)

for a, b in prior_parameters:
    plot_beta_posterior(a, b, s, n)
```
