### Problem

Let $Y_1,\,Y_2,\,\dots$ be independent random variables such that $P(Y_i = 1) = P(Y_i = -1) = 1/2$.
Let $X_n = \sum_{i=1}^n Y_i$.
Think of $Y_i = 1$ as "the stock price increased by one dollar",
$Y_i = -1$ as "the stock price decereased by one dollar",
and $X_n$ as the value of the stock on day $n$.
1. Find $\mathbb{E} (X_n)$ and $\mathbb{V} (X_n)$.
2. Simulate $X_n$ and plot $X_n$ versus $n$ for $n=1,\,2,\,\dots,\,10,000$.
   Repeat the whole simulation several times.
   Notice two things.
   First, it's easy to "see" patterns in the sequence even though it is random.
   Second, you will find that the four runs look very different even though they are generated the same way.
   How do the calculations in part 1 explain the second observation?

### Solution
Since $Y_i$ has mean zero we deduce that $\mathbb{E}(X_n) = 0$. Similarly, since the $Y_i$'s are independent with variance given by $\mathbb{V}(Y_i) = \mathbb{E}(Y_i^2) = 1$ we deduce that $\mathbb{V}(X_n) = \sum_{i=1}^n \mathbb{V} (Y_i) = n.$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli
```

```python
def plot_1D_random_walk(n: int, k=1):

    Y = [2*bernoulli.rvs(p=1/2, size=n) - 1 for _ in range(k)]
    X = [np.cumsum(y) for y in Y]

    npoints = np.arange(n)

    for j in range(k):
        plt.plot(npoints, X[j])

    plt.show()
```

```python
plot_1D_random_walk(10**6, 5)
```
