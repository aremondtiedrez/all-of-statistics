Let $V = (X, Y, Z)$ have the following joint distribution
$$
    X \sim \text{Bernoulli} \left( \frac{1}{2} \right)\\
    Y \sim \text{Bernoulli} \left( \frac{e^{4x-2}}{1+e^{4x-2}} \right)\\
    Z \sim \text{Bernoulli} \left( \frac{e^{2(x+y)-2}}{1+e^{2(x+y)-2}} \right)
$$
1. Find an expression for $\mathbb{P}(Z=z|Y=y)$.
   In particular find $\mathbb{P}(Z=1|Y=1)$.
2. Write a program to simulate the model.
   Conduct a simulation and compute $\mathbb{P}(Z=1|Y=1)$ empirically.
   Plot this as a function of the simulation size $N$.
   It should converge to the theoretical value you computed in (1).
3. Write down an expression for $\mathbb{P}(Z=z|Y:=1)$.
   In particular find $\mathbb{P}(Z=1|Y:=1)$.
4. Write a program to simulate the intervention "set $Y := 1$".
   Conduct a simulation and compute $\mathbb{P}(Z=1|Y:=1)$ empirically.
   Plot this as a function of the simulation size $N$.
   It should converge to the theoretical value you computed in (3).

```python
from collections import namedtuple
from scipy.special import expit # Logistic function

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

2. We define the empirical *counts*
$$
    N_{jk}
    := \# \{ i : Y_i = j, Z_i = k \}
    = \sum_{i=1}^n 1(Y_i = j, Z_i = k)
$$
and
$$
    N_{j\cdot} = \sum_{k=0}^1 N_{jk}
$$
such that
$$
    \hat{\mathbb{P}} (Z = k | Y = j) = \frac{N_{jk}}{N_{j\cdot}}.
$$
4. In this case we define the empirical *intervention* counts
$$
    M_{lk}
    := \# \{ i : X_i = l, Z_i = k \}
    = \sum_{i=1}^n 1(X_i = l, Z_i = k)
$$
and
$$
    \hat{\mathbb{P}} (Z = k | Y := 1) = \frac{M_{\cdot k}}{n} = \frac{1}{n} \sum_{l=0}^1 M_{lk}.
$$

```python
x = scipy.stats.bernoulli.rvs(p = 0.5, size = 1000)
y = np.full(shape=1000, fill_value=1)
z = scipy.stats.bernoulli.rvs(p = expit(2*(x + y) - 2))
```

```python
Data = namedtuple('Data', ['n', 'intervention', 'x', 'y', 'z'])
Estimate = namedtuple('Estimate', ['n', 'intervention', 'cond_prob'])

def generate_data(n, intervention):

    x = scipy.stats.bernoulli.rvs(p = 0.5, size = n)
    if intervention:
        y = np.full(shape=n, fill_value=1)
    else:
        y = scipy.stats.bernoulli.rvs(p = expit(4*x - 2))
    z = scipy.stats.bernoulli.rvs(p = expit(2*(x + y) - 2))
    
    return Data(n, intervention, x, y, z)

def estimator(data):

    if data.intervention:
        counts = np.array([(data.z == k).sum() for k in (0, 1)])
        active_conditioning_probabilities = counts/data.n
        return Estimate(data.n, data.intervention, active_conditioning_probabilities)
    else:
        counts = np.array(
            [[
                (np.stack((data.y, data.z), axis=1) == [j, k]).all(axis=1).sum()
            for k in (0, 1)] for j in (0, 1)]
        )
        passive_conditional_probabilities = np.array(
            [[
                counts[j, k]/counts.sum(axis=1)[j]
            for k in (0, 1)] for j in (0, 1)]
        )   
        return Estimate(data.n, data.intervention, passive_conditional_probabilities)

def print_report(estimate):
    
    if estimate.intervention:
        print(
            "Estimates of the active conditional probabilities.\n"
            "---------------------------------------------------\n"
            f"P(Z = 0 | Y := 1): {estimate.cond_prob[0]:.3}\n"
            f"P(Z = 1 | Y := 1): {estimate.cond_prob[1]:.3} <-- We focus on this quantity and call it lambda.\n"
        )
    else:
        print(
            "Estimates of the passive conditional probabilities.\n"
            "---------------------------------------------------\n"
            f"P(Z = 0 | Y = 0): {estimate.cond_prob[0, 0]:.3}\n"
            f"P(Z = 0 | Y = 1): {estimate.cond_prob[1, 0]:.3}\n"
            f"P(Z = 1 | Y = 0): {estimate.cond_prob[0, 1]:.3}\n"
            f"P(Z = 1 | Y = 1): {estimate.cond_prob[1, 1]:.3} <-- We focus on this quantity and call it theta.\n"
        )
```

```python
n=int(1e6)

print_report(estimator(generate_data(n, intervention=False)))
print_report(estimator(generate_data(n, intervention=True)))
```

```python
# Estimate theta for various values of n,
# then plot the results, comparing to the true value of theta
n_arr = np.logspace(start=1, stop=6, dtype='i')
theta_arr = [estimator(generate_data(n, intervention=False)).cond_prob[1, 1] for n in n_arr]
lambda_arr = [estimator(generate_data(n, intervention=True)).cond_prob[1] for n in n_arr]

theta_true = expit(2)**2 + expit(-2)/2
lambda_true = 1/4 + expit(2)/2

plt.semilogx(n_arr, theta_arr, label='Theta estimate')
plt.semilogx(n_arr, np.full(n_arr.shape, theta_true), label='Theta')
plt.semilogx(n_arr, lambda_arr, label='Lambda estimate')
plt.semilogx(n_arr, np.full(n_arr.shape, lambda_true), label='Lambda')

plt.legend(loc='best')
plt.show()
```
